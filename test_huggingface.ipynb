```python
import os
from dotenv import load_dotenv
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from langchain_core.prompts import ChatPromptTemplate
```

```python
# Load environment variables from .env file
load_dotenv()
```

```python

SYSTEM_PROMPT = """\
You are an up-and-coming yet sarcastic journalist. You have been assigned
to write an article about a shark attack. The article should take the short
description given and expand it into a longer article. The article should be
at most 200 words long, and at least 100 words long. Your goal is to make
cheeky jokes about the situation. All situations will be shark attacks.
"""

human_prompt = "Short description: {description}\n\nArticle:"

prompt = ChatPromptTemplate.from_messages([("system", SYSTEM_PROMPT), ("human", human_prompt),])

endpoint = HuggingFaceEndpoint(
    repo_id="mistralai/Mistral-7B-Instruct-v0.3",
    temperature=0.7,
    max_new_tokens=250,
    huggingfacehub_api_token=os.getenv("HUGGINGFACEHUB_API_TOKEN"),
)

llm = ChatHuggingFace(llm=endpoint)

chain = prompt | llm
```

```python
description = "A swimmer was bitten by a shark while paddleboarding off the coast of Florida."
response = chain.invoke({"description": description})
print(response.content)
```
